```python
import yfinance as yf
import pandas as pd
import numpy as np

# 1. Final Refined Ticker List (Removing UNI7-USD as requested)
tickers = [
    'BTC-USD', 'ETH-USD', 'BNB-USD', 'XRP-USD', 'SOL-USD', 'TRX-USD',
    'DOGE-USD', 'ADA-USD', 'BCH-USD', 'XMR-USD', 'LINK-USD', 'XLM-USD',
    'LTC-USD', 'AVAX-USD', 'HBAR-USD', 'SHIB-USD', 'CRO-USD',
    'DOT-USD', 'UNI-USD', 'AAVE-USD', 'OKB-USD', 'NEAR-USD', 'ETC-USD'
]

# 2. Fetch data from 01/01/2021 to 31/12/2025
# Note: end date is exclusive in yfinance, so we use 2026-01-01 to include Dec 31
raw_data = yf.download(
    tickers,
    start="2021-01-01",
    end="2026-01-01",
    interval="1d",
    auto_adjust=True
)['Close']

# 3. MBA Data Cleaning: 5-Year Date Range Validation
# We want to ensure the coins existed for the majority of this specific window
# Total expected days is roughly 1826 (365*5 + 1 leap day)
min_days = 1826 * 0.90
valid_tickers = raw_data.columns[raw_data.count() >= min_days]
five_year_data = raw_data[valid_tickers].dropna()

print(f"--- Processed Dataset (2021-2025) ---")
print(f"Final Coin Count: {len(five_year_data.columns)}")
print(f"Time Range: {five_year_data.index.min().date()} to {five_year_data.index.max().date()}")

# 4. Calculate Log Prices and Log Returns
# Adding +1e-9 (epsilon) to avoid log(0) errors with SHIB
log_price = np.log(five_year_data + 1e-9)
log_return = log_price.diff().dropna()

# 5. Remove any extreme outliers (Winsorization)
# Replacing any residual 'inf' with NaN and dropping them
log_return.replace([np.inf, -np.inf], np.nan, inplace=True)
log_return = log_return.dropna()

print(f"--- Processed 5-Year Dataset ---")
print(f"Final Coin Count: {len(log_return.columns)}")
print(f"Time Range: {log_return.index.min().date()} to {log_return.index.max().date()}")
```

```python
correlation_matrix = log_return.corr()
print("Correlation Matrix calculated successfully.")
```

```python
correlation_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)): # Start from i + 1 to avoid self-correlation and duplicate pairs
        crypto1 = correlation_matrix.columns[i]
        crypto2 = correlation_matrix.columns[j]
        correlation_value = correlation_matrix.iloc[i, j]

        if correlation_value > 0: # Only consider positive correlations
            # Sort names alphabetically to ensure unique pairs regardless of order
            if crypto1 < crypto2:
                correlation_pairs.append((crypto1, crypto2, correlation_value))
            else:
                correlation_pairs.append((crypto2, crypto1, correlation_value))

# Sort the list of tuples by correlation value in descending order
sorted_correlations = sorted(correlation_pairs, key=lambda x: x[2], reverse=True)

print(f"Extracted {len(sorted_correlations)} unique positive correlation pairs.")
print("Top 5 correlated pairs:")
for pair in sorted_correlations[:5]:
    print(f"  {pair[0]} - {pair[1]}: {pair[2]:.4f}")
```

```python
non_repeating_pairs = []
selected_cryptos = set()

for crypto1, crypto2, correlation in sorted_correlations:
    if crypto1 not in selected_cryptos and crypto2 not in selected_cryptos:
        non_repeating_pairs.append((crypto1, crypto2, correlation))
        selected_cryptos.add(crypto1)
        selected_cryptos.add(crypto2)

print(f"Found {len(non_repeating_pairs)} non-repeating cryptocurrency pairs.")
print("Non-repeating pairs (sorted by correlation, highest to lowest):")
for pair in non_repeating_pairs:
    print(f"  {pair[0]} - {pair[1]}: {pair[2]:.4f}")
```

```python
unique_coins_set = set()

# Select the first 5 non-repeating pairs
for crypto1, crypto2, _ in non_repeating_pairs[:5]:
    unique_coins_set.add(crypto1)
    unique_coins_set.add(crypto2)

# Convert the set to a list
unique_coins = list(unique_coins_set)

print(f"Extracted {len(unique_coins)} unique cryptocurrencies from the top 5 non-repeating pairs:")
print(unique_coins)
```

```python
from statsmodels.tsa.stattools import adfuller

print("Imported adfuller for ADF test.")
```

```python
adf_results = {} # Dictionary to store ADF test results

# Filter log_price for the unique coins
log_price_selected = log_price[unique_coins]

for coin in unique_coins:
    # Extract the log price series for the current coin
    series = log_price_selected[coin].dropna() # Ensure no NaNs in the series for ADF test

    if len(series) > 0: # Only perform test if series is not empty
        # Perform ADF test
        adf_test = adfuller(series, autolag='AIC')

        # Store relevant results
        adf_results[coin] = {
            'ADF Statistic': adf_test[0],
            'p-value': adf_test[1],
            'Critical Values': adf_test[4] # This is a dictionary of critical values
        }
    else:
        adf_results[coin] = {
            'ADF Statistic': np.nan,
            'p-value': np.nan,
            'Critical Values': np.nan
        }

# Convert results to a DataFrame for better readability
adf_df = pd.DataFrame.from_dict(adf_results, orient='index')

# Add a 'Stationarity' column based on p-value <= 0.05
adf_df['Stationarity'] = adf_df['p-value'].apply(lambda p: 'Stationary' if p <= 0.05 else 'Non-Stationary')

print("ADF Test Results for Log Prices of Selected Cryptocurrencies:")
display(adf_df)
```

## Confirm I(1) Property of Log Prices

### Subtask:
For the cryptocurrencies in the `cointegration_pairs` list, verify that their log prices are integrated of order 1 (I(1)). This is done by confirming that their first differences (log returns) are stationary. We will leverage the results from the `adf_df_returns_filtered` DataFrame generated earlier, which contains ADF test results for log returns, to confirm this prerequisite for the Engle-Granger test.


**Reasoning**:
First, I'll extract all unique cryptocurrency tickers from the `cointegration_pairs` list to identify the specific coins relevant for this check. This is necessary to filter the ADF test results for log returns.



```python
unique_coins_for_cointegration = set()
for crypto1, crypto2 in cointegration_pairs:
    unique_coins_for_cointegration.add(crypto1)
    unique_coins_for_cointegration.add(crypto2)

# Convert set to a list
unique_coins_for_cointegration_list = list(unique_coins_for_cointegration)

print(f"Unique cryptocurrencies identified from cointegration pairs: {unique_coins_for_cointegration_list}")
```

**Reasoning**:
Now that the unique cryptocurrencies from the cointegration pairs have been identified, I will filter the `adf_df_returns_filtered` DataFrame to show only the ADF test results for their log returns. This will allow us to confirm if their log returns are stationary, which implies their log prices are I(1).



```python
filtered_returns_stationarity = adf_df_returns_filtered.loc[unique_coins_for_cointegration_list]

print("ADF Test Results for Log Returns of Cryptocurrencies in Cointegration Pairs:")
display(filtered_returns_stationarity)

# Confirm all are stationary
all_stationary = filtered_returns_stationarity['Stationarity'].eq('Stationary').all()

if all_stationary:
    print("\nAll cryptocurrencies in cointegration pairs have stationary log returns (I(1)).")
else:
    print("\nWARNING: Not all cryptocurrencies in cointegration pairs have stationary log returns.")
```

```python
from statsmodels.formula.api import ols
from statsmodels.tsa.stattools import adfuller

# Initialize a dictionary to store Engle-Granger test results
engle_granger_results = {}

# Filter log_price to include only the relevant coins for cointegration testing
# This ensures that we are only working with the log prices of the coins identified as non-stationary in their log prices
# and part of the cointegration_pairs.

# First, gather all unique coins from cointegration_pairs
all_coins_in_cointegration_pairs = list(unique_coins_for_cointegration)
log_price_for_cointegration = log_price[all_coins_in_cointegration_pairs]

print("Starting Engle-Granger Cointegration Test for selected pairs...")

for crypto1, crypto2 in cointegration_pairs:
    # 1. Align their log price series for the current pair
    # Create a temporary DataFrame for the current pair, ensuring no NaNs
    df_pair = log_price_for_cointegration[[crypto1, crypto2]].dropna()

    if df_pair.empty:
        print(f"Warning: No overlapping data for {crypto1} and {crypto2}. Skipping.")
        engle_granger_results[(crypto1, crypto2)] = {'ADF Statistic': np.nan, 'p-value': np.nan, 'Critical Values': np.nan, 'Cointegrated': False}
        continue

    # Rename columns to be compatible with patsy/ols formula (replace hyphens with underscores)
    df_pair_renamed = df_pair.copy()
    renamed_crypto1 = crypto1.replace('-', '_')
    renamed_crypto2 = crypto2.replace('-', '_')
    df_pair_renamed.rename(columns={crypto1: renamed_crypto1, crypto2: renamed_crypto2}, inplace=True)

    # 2. Perform OLS regression
    # Regress crypto1's log price on crypto2's log price using renamed columns
    model = ols(f'{renamed_crypto1} ~ {renamed_crypto2}', data=df_pair_renamed).fit()

    # 3. Extract the residuals from the OLS regression
    residuals = model.resid

    # 4. Apply the Augmented Dickey-Fuller (ADF) test to these residuals
    adf_test_residuals = adfuller(residuals, autolag='AIC')

    # 5. Store the results
    p_value_residuals = adf_test_residuals[1]
    is_cointegrated = p_value_residuals <= 0.05 # Check for stationarity of residuals

    engle_granger_results[(crypto1, crypto2)] = {
        'ADF Statistic': adf_test_residuals[0],
        'p-value': p_value_residuals,
        'Critical Values': adf_test_residuals[4],
        'Cointegrated': is_cointegrated
    }

print("Engle-Granger Cointegration Test complete. Results stored in 'engle_granger_results'.")
```

```python
engle_granger_df = pd.DataFrame.from_dict(engle_granger_results, orient='index')
engle_granger_df.index.names = ['Crypto1', 'Crypto2']

print("Engle-Granger Cointegration Test Results:")
display(engle_granger_df)

# Filter to show only cointegrated pairs
cointegrated_pairs_df = engle_granger_df[engle_granger_df['Cointegrated'] == True]

print("\nCointegrated Pairs (based on p-value <= 0.05 for residuals):")
display(cointegrated_pairs_df)
```

```python
granger_results = {}

# Max lags to consider for the Granger Causality test
# A common choice is 1 to 5, given our daily data, 5 lags seems reasonable.
# We will use 5 lags as a starting point. Adjust if necessary based on data characteristics.
max_lags = 5

print(f"Performing Granger Causality tests for {len(granger_causality_pairs)} pairs with max_lags={max_lags}...")

for crypto_y, crypto_x in granger_causality_pairs:
    # Prepare the data for Granger causality test
    # The function expects a DataFrame with the dependent variable as the first column
    # and the independent variable as the second column.
    data_for_granger = log_return[[crypto_y, crypto_x]].dropna()

    if data_for_granger.empty:
        print(f"Warning: No overlapping data for {crypto_y} and {crypto_x}. Skipping pair.")
        continue

    # Test direction 1: Does X Granger-cause Y?
    # H0: X does not Granger-cause Y
    # The function returns a dictionary of test results for each lag
    test_result_x_to_y = grangercausalitytests(data_for_granger[[crypto_y, crypto_x]], max_lags, verbose=False)
    # Get the p-value for the F-test at the max_lags
    p_value_x_to_y = test_result_x_to_y[max_lags][0]['ssr_ftest'][1] # F-test p-value

    # Test direction 2: Does Y Granger-cause X?
    # H0: Y does not Granger-cause X
    data_for_granger_swapped = log_return[[crypto_x, crypto_y]].dropna()
    test_result_y_to_x = grangercausalitytests(data_for_granger_swapped[[crypto_x, crypto_y]], max_lags, verbose=False)
    p_value_y_to_x = test_result_y_to_x[max_lags][0]['ssr_ftest'][1] # F-test p-value

    granger_results[f'{crypto_x} -> {crypto_y}'] = p_value_x_to_y
    granger_results[f'{crypto_y} -> {crypto_x}'] = p_value_y_to_x

print("Granger Causality tests complete. Results stored in 'granger_results'.")

# Display the results for quick inspection
for test, p_value in granger_results.items():
    print(f"{test}: p-value = {p_value:.4f}")
```
